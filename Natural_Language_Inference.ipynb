```
## NLI | Model Name: 'facebook/bart-large-mnli' 
## Language: English | Model Architecture: BART
## Labels: 0: contradiction, 1: neutral, 2: entailment

from transformers import pipeline

path = 'facebook/bart-large-mnli'

classify = pipeline("sentiment-analysis", model=path, tokenizer=path)

#label_0 = contradiction, label_1 = neutral, label_2 = entailment
classify('Market surveillance in FICC has undergone, and continues to undergo, significant change as a result of regulation, the evolution of market structure, and technological developments. This Spotlight Review considers these structural and technological changes, the emergence of machine learning trading strategies, and sets out some of the challenges associated with these developments for surveillance teams in FICC markets')
```

```
## NLI | Model Name: 'roberta-large-mnli' 
## Language: English | Model Architecture: RoBERTa
## Labels: 0: CONTRADICTION, 1: NEUTRAL, 2: ENTAILMENT

from transformers import pipeline

path = 'roberta-large-mnli'

classify = pipeline("sentiment-analysis", model=path, tokenizer=path)

#label_0 = contradiction, label_1 = neutral, label_2 = entailment
classify('Market surveillance in FICC has undergone, and continues to undergo, significant change as a result of regulation, the evolution of market structure, and technological developments. This Spotlight Review considers these structural and technological changes, the emergence of machine learning trading strategies, and sets out some of the challenges associated with these developments for surveillance teams in FICC markets')
```
